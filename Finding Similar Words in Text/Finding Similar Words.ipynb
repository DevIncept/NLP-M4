```python
import re
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
texts = ['Do you carry a purse or bag with you?', 'The young man found the artifacts, this is how professionals discover!', 
        'He earned A grade on 90% score.', 'She wins the match after the gain of last point.', 'Search for examples of how to do things.']
```

```python
def similar_words(texts):
    stop_words = set(stopwords.words('english'))
    lemmatizer = WordNetLemmatizer()
    for text in texts:
        synonyms_array = {}
        text = re.sub(r'[^a-zA-Z\s]+', '', text)
        text = text.lower()
        word_tokens = word_tokenize(text)
        filtered_text = []
        for word in word_tokens:
            if word not in stop_words:
                filtered_text.append(word)
    #     print(word_tokens)
    #     print(filtered_text)
        lemmatized_words = []
        for word in filtered_text:
            lemmatized_words.append(lemmatizer.lemmatize(word))
    #     print(lemmatized_words)
        for word in lemmatized_words:
            synArr = []
            sArr = wordnet.synsets(word)
            for syn in sArr:
                synArr.append(syn.name()[:-5])
            synArr = list(set(synArr))
            for synonyms in synArr:
                if synonyms in lemmatized_words:
                    if synonyms != word:
                        synonyms_array[word] = synonyms
        if bool(synonyms_array):
            print(synonyms_array)
        else: 
            print('No synonyms found!')
```

```python
similar_words(texts)
```
